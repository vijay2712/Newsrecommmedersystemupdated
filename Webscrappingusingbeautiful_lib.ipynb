```python
import requests
import json
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import csv
```

```python
response = requests.get("https://www.dawn.com/latest-news")
soup = BeautifulSoup(response.text, 'lxml')
```

```python
print(soup.prettify())
```

```python
t=soup.find_all(class_="story__link")
title=[]
for tt in t:
    title.append(tt.text)
title=pd.DataFrame(title)
title=title.rename(columns={0:"Title"})
title
```

```python
t=soup.find_all(class_="story__excerpt font-georgia pb-1 px-0 sm:px-2 mt-0")
content=[]
for tt in t:
    content.append(tt.text)
content=pd.DataFrame(content)
content=content.rename(columns={0:"Content"})
content
```

```python
t=soup.find_all(class_="timestamp--time timeago")
Date_time=[]
for tt in t:
    Date_time.append(tt.text)
Date_time=pd.DataFrame(Date_time)
Date_time=Date_time.rename(columns={0:"Date_Time"})
Date_time
```

```python
Table=pd.concat([Date_time,title,content,],axis=1)
Table
```

```python
Table.info()
```

```python
Table=Table.dropna()
Table
```

```python
Table.to_csv('Table1.csv',index=False)
```
